**Praktikum 5 Visualisasi Data - Scatter Plot**

**1.Import Liblary**

```python
import matplotlib.pyplot as plt
```

**2.Buat Dataset**

```python
suhu = [20,22,24,26,28,30,32,34,36]
penjualan = [50,60,70,80,90,100,110,130,150]
```

**3.Basic Scatter Plot**

```python
import matplotlib.pyplot as plt
plt.scatter(suhu,penjualan)
plt.title('Hubungan penjualan Es krim dengan suhu ')
plt.xlabel('Suhu')
plt.ylabel('Penjualan Es krim')
plt.show()
```

**4.Kustomisasi Scatter Plot**

```python
plt.scatter(suhu,penjualan,color='red',marker='o',s=100)
plt.title('Hubungan penjualan Es krim dengan suhu ')
plt.xlabel('Suhu')
plt.ylabel('Penjualan Es krim')
plt.show()
```

**5.Multiple Scatter Plot**

```python
# Dataset suhu
Suhu = [20,22,24,26,28,30,32,34,36]

#Dataset Penjulaan
Penjualan_weekdays = [50,60,70,80,90,100,110,130,150]
Penjualan_weekend  = [60,70,80,90,100,110,130,150,190]

plt.scatter(Suhu,Penjualan_weekdays,color='red',s=100,label='Weekdays')
plt.scatter(Suhu,Penjualan_weekend,color='blue',s=100,label='Weekend')
plt.title('Hubungan penjualan Es krim dengan suhu ')
plt.xlabel('Suhu')
plt.ylabel('Penjualan Es krim')
plt.legend()
plt.show()
```

**6.Melihat Hubungan Data dengan Beberapa Variabel**

```python
# dataset
suhu = [20,22,24,26,28,30,32,34,36]
penjualan = [50,60,70,80,90,100,110,130,150]

# dataset tambahan
kelembapan = [60,65,70,75,80,85,90,95,100]
ukuran_titik = [200,220,240,260,280,300,320,340,360,]

#kustomisasi scatter plot
plt.scatter(suhu,penjualan,c=kelembapan,s=ukuran_titik,alpha=0.5,cmap='coolwarm')
plt.colorbar()
plt.title('Hubungan penjualan Es krim dengan suhu ')
plt.xlabel('Suhu')
plt.ylabel('Penjualan Es krim')
plt.show()
```

**7. 3D Scatter Plot**

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# dataset suhu
suhu = [20,22,24,26,28,30,32,34,36]

# dataset penjualan
penjualan_weekdays = [50,60,70,80,90,100,110,130,15]
penjualan_weekend = [60,70,80,90,100,110,120,140,160]

# 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(suhu,penjualan_weekdays,penjualan_weekend,c='r',marker='o')
ax.set_title('Hubungan Penjualan dengan Suhu')
ax.set_xlabel('Suhu')
ax.set_ylabel('Penjualan Weekdays')
ax.set_zlabel('Penjualan Weekend')
plt.show()
```
